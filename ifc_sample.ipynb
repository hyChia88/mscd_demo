```python
import json
import ifcopenshell
import ifcopenshell.util.element
import ifcopenshell.util.pset
import ifcopenshell.geom
from pathlib import Path
```

```python
ifc_path = "sample/Architectural.ifc"
model = ifcopenshell.open(ifc_path)

print("Schema:", model.schema)
```

```python

items = []

for e in model:
    ent = {
        "id": e.id(),                     # IFC line id: #123
        "type": e.is_a(),                 # e.g. IfcWall, IfcDoor
        "name": getattr(e, "Name", None),
        "global_id": getattr(e, "GlobalId", None),
    }
    items.append(ent)

print(json.dumps(items, indent=2))
```

```python
# ---- only building elements ----
building_elems = []
for el in model.by_type("IfcBuildingElement"):
    data = {
        "id": el.id(),
        "type": el.is_a(),
        "name": getattr(el, "Name", None),
        "global_id": getattr(el, "GlobalId", None),
        "psets": ifcopenshell.util.element.get_psets(el),
    }
    building_elems.append(data)

print("Building elements:", len(building_elems))
print(json.dumps(building_elems[:10], indent=2))  # show first 10
```
